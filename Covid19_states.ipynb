```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import urllib.request as url
import json
```

```python
response = url.urlopen('https://api.covid19india.org/states_daily.json')
```

```python
response
```

```python
json_data = json.load(response)
```

```python
json_data
```

```python
states_data = json_data['states_daily']
```

```python
df = pd.DataFrame(states_data)
```

```python
df
```

```python
df.shape
```

```python
df['rj']
```

```python
df.iloc[10:50]['rj']
```

```python
df.iloc[10:50][['mh','dl','gj','rj']]
```

```python
pd.unique(df['status'])
```

```python
conf = df[df['status'] == 'Confirmed']
```

```python
conf.shape
```

```python
conf['rj'] = conf['rj'].astype(int)
```

```python
conf['rj'].sum()
```

```python
plt.figure(figsize=(10,6))
conf['rj'].plot()
# plt.xlabel('')
# plt.ylabel('')
plt.title('Cases In Rajasthan')
```

```python
conf['date'] = pd.to_datetime(conf['date'])
```

```python
conf.head()
```

```python
plt.figure(figsize=(20,10))
plt.plot(conf['date'], conf['rj'])
plt.xlabel("Date")
plt.ylabel("Cases")
plt.title("Confirmed Cases in Rajasthan")
plt.xticks(rotation = 'vertical')
plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.plot(conf['date'], conf['rj'], label = 'Rajasthan')
plt.plot(conf['date'], conf['gj'], label = 'Gujrat')
plt.plot(conf['date'], conf['dl'], label = 'Delhi')
plt.xlabel("Date")
plt.ylabel("Cases")
plt.title("Confirmed Cases in Delhi, Gujrat & Rajasthan")
plt.xticks(rotation = 'vertical')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.bar(conf['date'],conf['rj'])
```
